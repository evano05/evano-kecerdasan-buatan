```python
import json

def get_sorted_companies(company_list):
    # Sort perusahaan berdasarkan nama dalam urutan terbalik
    sorted_companies = sorted(company_list, key=lambda x: x['name'], reverse=True)
    
    # Format hasil hanya dengan nama perusahaan
    result = [{"name": company["name"]} for company in sorted_companies]
    
    return result

# Data perusahaan
company_detail_list = [
    {'name': 'Company 1', 'domain': 'Retail', 'country': 'United States'},
    {'name': 'Company 2', 'domain': 'Technology', 'country': 'United Kingdom'},
    {'name': 'Company 3', 'domain': 'Healthcare', 'country': 'United States'}
]

# Jalankan fungsi dan tampilkan hasil dalam format JSON yang terstruktur
output = get_sorted_companies(company_detail_list)
print(json.dumps(output, indent=4))
```

```python
import json

def get_company_employee_nationality_count(company_list, employee_list):
    # Buat dictionary untuk menyimpan jumlah kebangsaan karyawan berdasarkan perusahaan
    company_nationality_count = {company['name']: {} for company in company_list}

    # Hitung jumlah kebangsaan karyawan untuk setiap perusahaan
    for employee in employee_list:
        company_name = employee['company']
        nationality = employee['nationality']
        
        # Update jumlah kebangsaan untuk perusahaan yang sesuai
        if nationality in company_nationality_count[company_name]:
            company_nationality_count[company_name][nationality] += 1
        else:
            company_nationality_count[company_name][nationality] = 1
    
    # Format hasil sesuai dengan struktur yang diinginkan
    result = [
        {
            "company": company,
            "employee_nationality": company_nationality_count[company]
        }
        for company in company_nationality_count
    ]
    
    return result

# Data
company_detail_list = [
    {'name': 'Company 1', 'domain': 'Retail', 'country': 'United States'},
    {'name': 'Company 2', 'domain': 'Technology', 'country': 'United Kingdom'},
    {'name': 'Company 3', 'domain': 'Healthcare', 'country': 'United States'}
]

employee_detail_list = [
    {'name': 'EMP-0001', 'first_name': 'John', 'last_name': 'Doe', 'full_name': 'John Doe', 'company': 'Company 1', 'nationality': 'Australia'},
    {'name': 'EMP-0002', 'first_name': 'Tom', 'last_name': 'Smith', 'full_name': 'Tom Smith', 'company': 'Company 2', 'nationality': 'United States'},
    {'name': 'EMP-0003', 'first_name': 'Andrew', 'last_name': 'Sebastian', 'full_name': 'Andrew Sebastian', 'company': 'Company 3', 'nationality': 'United States'},
    {'name': 'EMP-0005', 'first_name': 'Ying Han', 'last_name': 'Tan', 'full_name': 'Ying Han Tan', 'company': 'Company 1', 'nationality': 'Australia'},
    {'name': 'EMP-0015', 'first_name': 'Kenneth', 'last_name': 'Ng', 'full_name': 'Kenneth Ng', 'company': 'Company 3', 'nationality': 'United States'},
    {'name': 'EMP-0018', 'first_name': 'Rubby', 'last_name': 'Lee', 'full_name': 'Rubby Lee', 'company': 'Company 2', 'nationality': 'Hong Kong'},
    {'name': 'EMP-0017', 'first_name': 'Robert', 'last_name': 'White', 'full_name': 'Robert White', 'company': 'Company 1', 'nationality': 'United Kingdom'}
]

# Jalankan fungsi dan tampilkan hasil dalam format JSON yang terstruktur
output = get_company_employee_nationality_count(company_detail_list, employee_detail_list)
print(json.dumps(output, indent=4))
```
